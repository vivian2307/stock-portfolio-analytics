```python
# Cổ phiếu có volume cao nhất
total_volume = volume_df.sum()
max_volume_stock = total_volume.idxmax()
m_volume = total_volume[max_volume_stock]
pct_volume_max = (m_volume / total_volume.sum()) * 100

print(f"The highest volume symbol: {max_volume_stock}\n"
      f"Its value: {m_volume}\n"
      f"Its percentage of total volume: {pct_volume_max:.2f}%")
```

```python
# Cổ phiếu có volume thấp nhất
total_volume = volume_df.sum()
min_volume_stock = total_volume.idxmin()
min_volume = total_volume[min_volume_stock]
pct_volume_min = (min_volume / total_volume.sum()) * 100
print(f"The lowest volume symbol: {min_volume_stock}\n"
      f"Its value: {min_volume}\n"
      f"Its percentage of total volume: {pct_volume_min:.2f}%")
```

```python
# Biểu đồ boxplot của khối lượng giao dịch (Volume) có thể cung cấp cái nhìn về
# phân phối của khối lượng giao dịch của các cổ phiếu của các công ty.
import matplotlib.pyplot as plt
import seaborn as sns

# Chuẩn bị dữ liệu
volume_data = data['Volume']

# Vẽ boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=volume_data)
plt.title('Boxplot of Volume')
plt.ylabel('Volume')
plt.xlabel('Ticker')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mplfinance.original_flavor import candlestick_ohlc
import pandas as pd
import matplotlib.dates as mdates

# Tạo subplot
fig, ax = plt.subplots()

# Chuyển đổi cột 'Date' sang định dạng ngày mốc Matplotlib
ohlc_data = data.reset_index()
ohlc_data['Date'] = ohlc_data['Date'].map(mdates.date2num)

# Vẽ biểu đồ Candlestick
candlestick_ohlc(ax, ohlc_data.values, width=0.6, colorup='green', colordown='red')

# Đặt tiêu đề và nhãn trục
ax.set_title('Candlestick Chart')
ax.set_xlabel('Date')
ax.set_ylabel('Price')

# Định dạng lại ngày trên trục x
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Xoay nhãn trục x
plt.xticks(rotation=45)

# Hiển thị biểu đồ
plt.show()
```

```python
import seaborn as sns

volume_price_data = pd.DataFrame()

for price, ticker in data.columns:
    if price == 'Volume':
        volume_price_data['Volume'] = data[price, ticker]
    elif price in ['Adj Close']:
        volume_price_data[price + '_' + ticker] = data[price, ticker]

# Kiểm tra xem 'Volume' có trong DataFrame hay không
if 'Volume' in volume_price_data.columns:
    # Áp dụng regplot cho cột 'Volume' và 'Price' cho mỗi ticker
    for col in volume_price_data.columns:
        if 'Volume' in col:
            continue
        sns.regplot(x='Volume', y=col, data=volume_price_data, label=col)

    # Đặt tiêu đề và nhãn trục
    plt.title('Volume vs Price Regression')
    plt.xlabel('Volume')
    plt.ylabel('Price')
    plt.legend()

    # Hiển thị biểu đồ
    plt.show()
else:
    print("No 'Volume' column found in the DataFrame.")
```

```python
# Vẽ heatmap
plt.figure(figsize=(10, 8))  # Đặt kích thước của biểu đồ
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")  # Vẽ heatmap với các giá trị tương quan

# Đặt tiêu đề cho biểu đồ
plt.title('Correlation Heatmap')

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd

# Tạo cột 'Ticker' với giá trị 'AAPL', 'GOOG' và 'MSFT'
tickers = ['AAPL'] * len(data) + ['GOOG'] * len(data) + ['MSFT'] * len(data)

# Tạo cột 'Date_' bằng cách lặp lại cột 'Date_' gốc cho mỗi ticker
dates = data.index.tolist() * 3

# Tạo DataFrame mới với các cột 'Date_', 'Ticker', và các cột giá khác
tech_giant = pd.DataFrame({'Date_': dates, 'Ticker': tickers})

# Lặp lại các giá trị của các cột 'Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume' cho mỗi ticker
for price_type in ['Adj Close', 'Close', 'High', 'Low', 'Open', 'Volume']:
    # Tạo danh sách giá trị cho từng ticker
    values = []
    for ticker in ['AAPL', 'GOOG', 'MSFT']:
        values.extend(data[(price_type, ticker)].values)
    # Thêm cột giá trị mới vào DataFrame
    tech_giant[f'{price_type}'] = values

# Hiển thị DataFrame mới
print(tech_giant)
```
